# ICS3U Python Programming Evaluation Question 1

File Name:  `ICS3Ushape3D.py`

A designer is creating a structure that looks like this: 



- The cost of building this structure depends on its volume and surface area. 

- He needs a fast and easy way to calculate those volume and surface area values with different dimensions. 

- You need to write a program that helps him find those values for different inputs of a,b,c,d,e

Note:  The badly drawn arc in the center is circular.

```python
import math
a = float(input("Please enter the value of \"a\" in meters"))
b = float(input("Please enter the value of \"b\" in meters"))
c = float(input("Please enter the value of \"c\" in meters"))
d = float(input("Please enter the value of \"d\" in meters"))
e = float(input("Please enter the value of \"e\" in meters"))
radius = (e-a-b)/2
surface_area_of_circle = (math.pi*(math.pow(radius, 2))/2)
surfacearea_of_frontside = round(e*d - surface_area_of_circle, 2)
round_surface_area = c*math.pi*radius
surfacearea = round((e*c + 2*d*c + b*c + a*c + 2*surfacearea_of_frontside + round_surface_area), 2)
volume_of_cylinder = round((surface_area_of_circle*c), 2)
volume = round(e*d*c - volume_of_cylinder)
print("The surface area of the shape is",surfacearea,"meter squared and the volume of the shape is",volume,"meter cubed.")
```
